```python
import numpy as np

# This method is used to find a line(y = mx+c) which best represents a linear relation between the training dataset.

def gradient_descent(x,y):
    m_curr = 0
    c_curr = 0
    n = len(x)
    iterations = 10000
    learning_rate = 0.001
    
    for i in range(iterations):
        
        y_predicted = m_curr*x + c_curr
        cost = (1/n)*sum(val**2 for val in y-y_predicted)
        md = -(2/n)*sum(x*(y-y_predicted))               #md stands for partial derivative of cost function wrt m(slope)
        cd = -(2/n)*sum(y-y_predicted)                   #cd stands for partial derivative of cost function wrt c(intercept)
        m_curr = m_curr - (learning_rate*md)  
        c_curr = c_curr - (learning_rate*cd)
        
        print("interation: {}, m:{}, c:{},  cost function:{} ".format(i, m_curr, c_curr, cost))
        

x_input = np.array([1,2,3,4,5])
y_output = np.array([5,7,9,11,13])

#here you can see the relation could be reprented as y_output = 2(x_input) + 3 
# m=2, c=3
#Therefore, calling the gradient_descent function m---> 2, c---> 3 and cost---> 0

gradient_descent(x_input, y_output)
        
```
